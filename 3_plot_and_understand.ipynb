```python
import sklearn
```

```python
# for this we take the dataset load_boston. but it is not in colab for now. but they provide information that you need to add it to your leaning
from sklearn.datasets import load_boston
```

```python
#now we use the suggestion colab gave
import pandas as pd
import numpy as np
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
X = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
y = raw_df.values[1::2, 2]
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
mod_kn = KNeighborsRegressor()
```

```python
mod_kn.fit(X, y)
```

```python
pred_kn = mod_kn.predict(X)
```

```python
#now lets import matplot library for plotting the data. it is a framework/tool called 'matplotlib.pylab'. but since the name is long we import it wich short assigned name. generally it is 'plt'

import matplotlib.pylab as plt
```

```python
##lets plot a scatter plot of the predicted vales vs actual values using 'plt.scatter(x,y)'

plt.scatter(y, pred_kn)
```
