```python
import sklearn
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
fetch_california_housing()
```

```python
# now lets assign variables for propertis and targets

X, y = fetch_california_housing(return_X_y=True)
```

```python
# neighbour models are in ' neighbour's

from sklearn.neighbors import KNeighborsRegressor
```

```python
# now we need make a model object to use

mod = KNeighborsRegressor()
```

```python
# now lets say my model to make prection using the same X values.
# '<model object>.precitic(<Input>)' return the prediction. so that we need a variable to hold the return

## here we get NotFittedError. it means model has not been trained yet
## to train that we need to use '.fit(<Feature,Targets>)'


pred = mod.predict(X)
```

```python
#training the model
mod.fit(X, y)
```

```python
# now we can make predictions using our trained model object

pred = mod.predict(X)

print(pred)
```

```python
## one of the good thing in scikit learn is that we can work with defferent machine learning models with same api calls.
## lets enter a linear_model . its in 'linear_models'. then we can do the next steps as we did in the above example. same calls


from sklearn.linear_model import LinearRegression
```

```python
mod_linear = LinearRegression()
```

```python
mod_linear.fit(X, y)
```
